```python
import cPickle as pickle

from program_synthesis.datasets.karel import coverage
```

```python
f = open('../data/karel/val.pkl')
try:
    while True:
        item = pickle.load(f)
        cov = coverage.CoverageMeasurer(item['code'])
        for example in item['examples'][:5]:
            success = cov.add(example['in'])
            assert success
        stmt_missing, branch_missing = cov.uncovered()
        
        if stmt_missing or branch_missing:
            print 'Code: {}'.format(' '.join('{}:{}'.format(i, token) for i, token in enumerate(item['code'])))
            print 'Statements: {}'.format(', '.join(str(x) for x, y in sorted(stmt_missing)))
            print 'Conditionals: {}'.format(', '.join('{}-{} {}'.format(x, y, cond) for (x, y), cond in branch_missing))
            print
except EOFError:
    pass
            
```
